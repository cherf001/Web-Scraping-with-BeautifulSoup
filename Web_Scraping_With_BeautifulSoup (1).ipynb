```python
#get href with csv
links=bs.find_all(href = True)
csv_links=[]
for link in links:    # The for loop itirates through and returns the number of links in connection with the data.
    href=link.get('href')
    
    if href.find('.csv') != -1:
        csv_links.append(href)
print(len(csv_links))
```

```python
# Getting the links of the data
(csv_links)
```

```python
url=csv_links[2]
```

```python
# Loading the required libraries and the data.
import requests
import pandas as pd
import io
s=requests.get(url).content
data=io.StringIO(s.decode("utf-8"))
df=pd.read_csv(data)
df.head()
```
